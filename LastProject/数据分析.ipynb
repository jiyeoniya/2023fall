```python
average_salary = df.groupby('工作城市')['平均工资(K)'].mean()#平均工资
average_low_salary = df.groupby('工作城市')['最小薪资(K)'].mean()#最低平均工资
average_high_salary = df.groupby('工作城市')['最大薪资(K)'].mean()#最高平均工资
x = average_salary.reset_index()['工作城市'].tolist()
y1 = average_salary.reset_index()['平均工资(K)'].tolist()
y2 = average_low_salary.reset_index()['最小薪资(K)'].tolist()
y3 = average_high_salary.reset_index()['最大薪资(K)'].tolist()
print(x)
bar = Bar(init_opts=opts.InitOpts(width='1600px',height='600px',theme=ThemeType.MACARONS))
bar.add_xaxis(x)
bar.add_yaxis("各城市平均工资", y1)
bar.add_yaxis("各城市最低平均工资", y2)
bar.add_yaxis("各城市最高平均工资", y3)
bar.set_series_opts(label_opts=opts.LabelOpts(is_show=False))#不显示数字
bar.set_global_opts(toolbox_opts=opts.ToolboxOpts(is_show=True),
                    title_opts=opts.TitleOpts(title="各城市工资"),
                    yaxis_opts=opts.AxisOpts(name="薪资(K)"),
                    xaxis_opts=opts.AxisOpts(name="城市",axislabel_opts=opts.LabelOpts(rotate=-50))#旋转x轴内容
                   )
# datazoom_opts=opts.DataZoomOpts()#滚动条
bar.render("各城市工资.html")
```

```python
data_company = pd.read_csv('D:\桌面\lastProject\job_company.csv',encoding='GB2312')
data_company.columns=["","工作名称","工作城市","工作地点","工作公司","工作薪资","最大薪资(K)","最小薪资(K)","平均工资(K)"]
data_company
```

```python
company_ave_salary = data_company.groupby('工作公司')['平均工资(K)'].mean()
company_low_salary = data_company.groupby('工作公司')['最小薪资(K)'].mean()
company_high_salary = data_company.groupby('工作公司')['最大薪资(K)'].mean()


x = company_ave_salary.reset_index()['工作公司'].tolist()
print(x)
y1 = company_ave_salary.reset_index()['平均工资(K)'].tolist()
print(y1)
y2 = company_low_salary.reset_index()['最小薪资(K)'].tolist()
y3 = company_high_salary.reset_index()['最大薪资(K)'].tolist()
bar = Bar(init_opts=opts.InitOpts(width='800px',height='400px',theme=ThemeType.MACARONS))
bar.add_xaxis(x)
bar.add_yaxis("公司平均工资", y1,)# itemstyle_opts=opts.ItemStyleOpts(color="black")
bar.add_yaxis("公司最低平均工资", y2)
bar.add_yaxis("公司最高平均工资", y3)
bar.set_series_opts(label_opts=opts.LabelOpts(is_show=False))
bar.set_global_opts(toolbox_opts=opts.ToolboxOpts(is_show=True),
                    title_opts=opts.TitleOpts(title="大厂工资待遇"),
                    yaxis_opts=opts.AxisOpts(name="薪资(K)"),
                    xaxis_opts=opts.AxisOpts(name="公司"))
bar.render("大厂工资待遇.html")
```
